### funciÃ³n para guardar

```python
# Funcion para guardar los subset de datos
def savedata(data, name):
    data.to_csv(f'{name}.csv', index=False)
```

### ranking de popularidad

```python
most_popular_dogs = dogs_data.sort_values('ranking_de_popularidad')[['ranking_de_popularidad', 'raza']].head(10)
most_popular_dogs
```

```python
savedata(most_popular_dogs, 'most_popular_dogs')
```

```python
alt.Chart(most_popular_dogs).mark_bar().encode(
   alt.X('raza:N', sort='y', title="Raza"),
   alt.Y('calificacion_popular:Q', title="Calificacion Popular")
)
```

### numero de razas entre cada rango de longevidad

```python
longevidad_media = dogs_data.groupby('longevidad_media').size().to_frame(name = 'count').reset_index()
savedata(longevidad_media, 'longevidad_media')
```

### proporcion de gastos con gastos en comida

### gastos en comida

```python
gasto_comida_vida_df = dogs_data[['raza', 'costo_total_vida']]
gasto_comida_vida_df['tipo_de_gasto'] = 'comida'
gasto_comida_vida_df.rename(columns={'costo_total_vida' : 'valor_en_dolares'}, inplace=True)
gasto_comida_vida_df
```

### otros gastos

```python
otros_gastos_vida_df = dogs_data[['raza', 'prom_otros_gastos_vida']]
otros_gastos_vida_df['tipo_de_gasto'] = 'otros_gastos'
otros_gastos_vida_df.rename(columns={'prom_otros_gastos_vida' : 'valor_en_dolares'}, inplace=True)
otros_gastos_vida_df
```

### unamoslos!

```python
spends_per_life_df = pd.concat([gasto_comida_vida_df, otros_gastos_vida_df], ignore_index=True)
spends_per_life_df
savedata(spends_per_life_df, 'spends_per_life_df')
```
